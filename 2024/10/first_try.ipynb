```python
def format_input(dat: str):
    grid = [[int(x) for x in list(l)] for l in dat.splitlines()] 
    start_pos = [(x,y) for y,l in enumerate(grid) for x,c in enumerate(l) if c == 0]
    start_pos = sorted(start_pos, key=lambda x: (x[1], x[0]))
    return grid, start_pos
```

```python
def move(x, y, grid) -> int:

    w, h = len(grid[0]), len(grid)

    print('current: ', x, y)
    if grid[y][x] == 9:
        print('return 1')
        return 1

    next_paths = []
    for j in range(-1, 2):
        for i in range(-1, 2):
            xn, yn = x+i, y+j
            if (xn < w) and (yn < h) and ((grid[yn][xn] - grid[y][x]) == 1):
                next_paths.append((xn, yn))
    
    if len(next_paths) == 0:
            return 0

    
    print('next paths:', next_paths)
    print('----') 
    return sum([move(x, y, grid) for x,y in next_paths])
```
