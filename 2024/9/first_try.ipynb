```python
## conclusion
def format_input(dat: str) -> str:
    file_space = [dat[i] for i in range(len(dat)) if i % 2 == 0]
    inp = [{'id': str(i), 'space': int(file_space[i])} for i in range(len(file_space))]
    empty_space = [int(dat[i]) for i in range(len(dat)) if i % 2 == 1]
    rep = ''
    for i in range(len(empty_space)):
        rep += f"{('#' + inp[i]['id'] + '#')*inp[i]['space']}{'.'*empty_space[i]}"
    rep += f"{('#' + inp[-1]['id'] + '#')*inp[-1]['space']}"
    return rep


def find_rightmost_digit(inp: str) -> str:
    pat = r'\.*(#\d+#)'
    mt = re.match(pat, inp[::-1])
    dg = mt.group(1)[::-1]
    return dg


def remove_rightmost_digit(inp: str) -> str:
    pat = r'(#\d+#)'
    return re.sub(pat, '.', inp[::-1], count=1)[::-1]


def replace_leftmost_space(inp: str, dg: str) -> str:
    return re.sub(r'\.', dg, inp, count=1)


def move_one_digit(inp: str) -> str:
    dg = find_rightmost_digit(inp)
    inp = remove_rightmost_digit(inp)
    inp = replace_leftmost_space(inp, dg)
    return inp


def there_is_empty_space(inp: str) -> bool:
    #pattern = r'(#\d+#)+\.+(#\d+#)+'
    pattern = r'.*#\.+#'
    return bool(re.match(pattern, inp))


def move_all_digits(inp: str) -> str:
    while there_is_empty_space(inp):
        inp = move_one_digit(inp)
    return inp


def extract_digits(inp: str) -> list[int]:
    pat = r"#\d+#"
    return [int(re.sub('#', '', s)) for s in re.findall(pat, inp)]


def get_checksum(inp: str) -> int:
    digits = extract_digits(inp)
    return sum([i*d for i,d in enumerate(digits)])


def sol1(dat: str) -> int:
    inp = format_input(dat)
    inp = move_all_digits(inp)
    return get_checksum(inp)
    
```

```python
def move_one_file(fid: str, inp: str):
    fe = [e for e in inp if e["id"] == fid][0]
    eclst = [e for e in inp if e['id'] == '.' and e['space'] >= fe['space'] and e["pos"] < fe["pos"]]
    if len(eclst) > 0:
        ec = eclst[0]
        diff_space = ec["space"] - fe["space"]
        fe_pos_init = fe["pos"]
        fe["pos"] = ec["pos"]
        ec["space"] = diff_space
        inp.append({'id': '.', 'space': fe["space"], 'pos': fe_pos_init})
        incr = [ee for ee in inp if ee["pos"] >= fe["pos"] and ee["id"] != fe["id"]]
        for ee in incr: ee["pos"] += 1
    inp = sorted(inp, key = lambda x: x["pos"])
    return inp
    
```

```python
def concat_spaces(inp: list[dict]) -> list[dict]:
    out = []
    j = 0
    while j < len(inp):
        if inp[j]["id"] != '.':
            out.append(inp[j])
            j += 1
        else:
            sp = 0
            while j < len(inp) and inp[j]["id"] == '.':
                sp += inp[j]["space"]
                j += 1 
            if sp > 0: out.append({"id": ".", "space": sp})
    return out
```
