```python
d = {'<': (-1, 0), '>': (1, 0), '^': (0, -1), 'v': (0, 1)}
class Robot:
    def __init__(self, g, init):
        self.x = init[0]
        self.y = init[1]
        self.g = g
        self.w = len(self.g[0])
        self.h = len(self.g)
        self.boxes = self._get_boxes()
    
    def _get_boxes(self):
        return [((i,j),(i+1,j)) for j,l in enumerate(grid) for i,x in enumerate(l) if x == '[']
    
    def update_pos(self, m):
        dx, dy = d[m][0], d[m][1]
        grid[self.y][self.x] = '.'
        self.x, self.y = self.x + dx, self.y + dy
        grid[self.y][self.x] = '@'
    
    def get_box(self, xm, ym):
        bb = [b for b in self.boxes if any([(i==xm and j==ym) for (i,j) in b])]
        if len(bb) == 0:
            return None 
        return bb[0]

    def blocked(self, b, m):
        dx, dy = d[m][0], d[m][1]
        for c in b:
            xc, yc = c
            xm, ym = xc + dx, yc + dy
            if self.g[ym][xm] == '#':
                return True
        return False

    def move(self, m):
        #print(m)
        dx, dy = d[m][0], d[m][1]
        #print(dx, dy)
        xi, yi = (self.x+dx, self.y+dy)
        #print('initial:', xi, yi)
        xm, ym = xi, yi 
        emp = None
        t = grid[yi][xi]
        goods = []
        while t != '#' and emp is None and ym >= 0 and ym < self.h and xm >= 0 and xm < self.w: 
            if t == '.': emp = (xm, ym)
            if t in ["[", "]"]: goods.append(t)
            xm, ym = xm+dx, ym+dy
            t = grid[ym][xm]
        if emp is None:
            return 
        xe, ye = emp
        grid[ye][xe] = 'O'
        grid[yi][xi] = '@'
        grid[self.y][self.x] = '.'
        self.x, self.y = xi, yi        
    
    def gps(self):
        return sum([100*j+i for j,l in enumerate(self.g) for i,x in enumerate(l) if x == 'O'])
    
    def boxes_in_contact(self, b, m):
        dx, dy = d[m][0], d[m][1]
        cb1, cb2 = b[0], b[1]
        xb1, yb1 = cb1
        xb2, yb2 = cb2
        cm1 = xb1 + dx, yb1 + dy
        cm2 = xb2 + dx, yb2 + dy
        return [
            bm 
            for bm in self.boxes 
            if 
                (bm[0] in [cm1, cm2] or bm[1] in [cm1, cm2])
                and bm != b
        ]
```
